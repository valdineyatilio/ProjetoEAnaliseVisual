A idade influencia diretamente os níveis de colesterol e pressão arterial?

# Algoritmo

# Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Carregando o dataset

```python
url = "https://raw.githubusercontent.com/valdineyatilio/ProjetoEAnaliseVisual/main/cleaned_merged_heart_dataset.csv"
df = pd.read_csv(url, delimiter=",", encoding="utf-8")
```

# Exibindo as primeiras linhas do dataset

```python
print(df.head())
```

# Verificando as colunas disponíveis


```python
print(df.columns)
```

# Gráfico 1: Distribuição da Idade dos Pacientes
Este gráfico mostra a distribuição etária no dataset, 
permitindo identificar a faixa etária predominante.


```python
plt.figure(figsize=(10, 6))
sns.histplot(df["age"], bins=20, kde=True)
plt.title("Distribuição da Idade dos Pacientes")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

# Gráfico 2: Histograma da Pressão Arterial
Este gráfico mostra a distribuição da pressão arterial em repouso,
permitindo identificar padrões entre os pacientes

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["trestbps"], bins=20, kde=True)
plt.title("Distribuição da Pressão Arterial dos Pacientes")
plt.xlabel("Pressão Arterial de Repouso")
plt.ylabel("Frequência")
plt.grid(True)
plt.show()
```

# Gráfico de Dispersão (Idade vs Colesterol)
Este gráfico ajuda a visualizar a relação entre a idade e os níveis de colesterol dos pacientes.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["age"], y=df["chol"], hue=df["sex"], palette="coolwarm")
plt.title("Relação entre Idade e Níveis de Colesterol")
plt.xlabel("Idade")
plt.ylabel("Nível de Colesterol")
plt.grid(True)
plt.show()
```

# Gráfico 4: Mapa de Calor das Correlações entre Fatores de Risco
Este mapa de calor mostra as correlações entre múltiplas 
variáveis do dataset, ajudando a identificar padrões 
entre idade, pressão arterial, colesterol, etc.


```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação Entre Variáveis do Dataset")
plt.show()
```

# Narrativa: 
A Relação Entre Idade e Saúde Cardiovascular

Personagem:

Carlos, um homem de 55 anos, sempre foi ativo e saudável. 
No entanto, ao realizar exames de rotina, percebeu que seus níveis de colesterol e pressão arterial começaram a subir.
Ele se pergunta: "Será que minha idade está impactando diretamente minha saúde cardiovascular?"

Cenário:

O contexto dessa análise é a saúde cardiovascular, um tema crucial à medida que a população envelhece. 
O dataset escolhido contém informações sobre idade, pressão arterial e colesterol, permitindo explorar padrões entre diferentes faixas etárias.

Enredo:

Carlos percebeu mudanças nos seus exames e quer entender se isso é um fator natural do envelhecimento ou se há outras razões para essas alterações.  
Para responder a essa dúvida, analisamos um conjunto de dados de pacientes, observando como esses indicadores variam conforme a idade.

Conflito:

A dúvida de Carlos reflete um problema maior: Existe uma tendência clara de aumento do colesterol e pressão arterial com a idade? 
Se for verdade, quais medidas podem ser tomadas para prevenir complicações futuras?

Desenvolvimento:

Para investigar essa questão, realizamos análises utilizando Python e bibliotecas de visualização de dados. 
Foram geradas três visualizações principais:
- Histograma da Idade - Mostra a distribuição etária dos pacientes e identifica quais faixas etárias predominam.
- Gráfico de Dispersão (Idade vs. Colesterol) - Permite visualizar se há uma relação clara entre o aumento da idade e níveis de colesterol.
- Mapa de Calor das Correlações - Identifica padrões entre múltiplas variáveis, revelando se idade e pressão arterial estão fortemente associadas.

Clímax:

Após as análises, descobrimos uma tendência clara: conforme a idade avança, há um aumento significativo nos níveis de colesterol e pressão arterial. 
Isso indica que o envelhecimento realmente pode ser um fator de risco para doenças cardiovasculares.

Desfecho:

A conclusão aponta que, para mitigar esses efeitos, mudanças no estilo de vida são essenciais.
Carlos e qualquer pessoa preocupada com sua saúde devem considerar hábitos como alimentação equilibrada, prática regular de exercícios
e acompanhamento médico para manter os níveis de colesterol e pressão arterial sob controle.
